#### DETECTING OUTLIERS USING THE INTERQUANTILE RANGE (IQR)

```python
Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile (0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df[col] < lower_bound) | (df[col]> upper_bound)]
print (f'{col}, {len(outliers)} outliers')
```

##### It is necessary to work on the outliers after working on the skewness, because outliers can still distort the model training, so i detect them using  the interquantile range, and this shows that the longitude_GPS_sq has 3,445 outliers

```python
#TREATING OUTLIERS USING WINSORIZATION
df[col] = np.where(df[col] < lower_bound, lower_bound, np.where(df[col] > upper_bound, upper_bound, df[col]))
```

```python
#CONVERTING THE TARGET COLUMN FROM STRING TO INTEGER
df['Good_Bad_Flag'] = df['Good_Bad_Flag'].astype(int)
```

#### PROCESSING THE NUMERICAL AND CATEGORICAL COLUMN

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer, make_column_selector as selector
from sklearn.impute import SimpleImputer
```

```python
num_sel = selector(dtype_include = np.number)
cat_sel = selector(dtype_include = ['object', 'category', 'bool'])

num_tf = Pipeline([
    ('impute', SimpleImputer(strategy = 'median')),
    ('scale', StandardScaler(with_mean = True))
])
cat_tf = Pipeline([
    ('impute', SimpleImputer(strategy = 'most_frequent')),
    ('encode', OrdinalEncoder(handle_unknown = 'use_encoded_value', unknown_value = -1))
])
preprocessor = ColumnTransformer([
    ('num', num_tf, num_sel),
    ('cat', cat_tf, cat_sel)
], remainder = 'drop')
```

##### Preprocessing is done before feeding my data to models, and this is done to; autommatically handle missing values, ensure the model is robust and reduce bias caused by different scales or missing values

```python
results = []

fig, axes = plt.subplots(1, 3, figsize = (18, 5))
for name, model in models.items():
    if name in ["Logistic Regression", "Support Vector Machine", "K-Nearest Neighbors"]:
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
        y_prob = model.predict_proba(X_test_scaled)[:, 1]
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:, 1]

    results.append({
        "Model" : name,
        "Test Accuracy" : accuracy_score(y_test, y_pred),
        "Precision Score" : precision_score(y_test, y_pred, pos_label = 1),
        "Recall Score" : recall_score(y_test, y_pred, pos_label = 1),
        "F1-Score" : f1_score(y_test, y_pred, pos_label = 1),
        "ROC-AUC" : roc_auc_score(y_test, y_prob)
        
})

results_df = pd.DataFrame(results).sort_values(by = "ROC-AUC", ascending = False)
results_df
```
